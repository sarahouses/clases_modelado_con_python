### Regresión Lineal

```python
n=np.size(moscu)
y= moscu
x=fechas
```

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-(np.sum(x)**2))
b = np.mean(y)-a*np.mean(x)
```

```python
print(f"la pendiente es{a}\nla intercección es {b}")
```

```python
y=a*x+b
```

```python
y
```

```python
np.linalg.norm(moscu-y)
```

```python
plt.plot(fechas,moscu)
plt.plot(x,y)
plt.plot(25,a*25+b,"*r")
plt.plot(60,a*60+b,"*c")
```

### Valores Aleatorios

```python
#float
np.random.randint(1,20,2)
```

```python
#decimal
np.random.uniform(1,20,2)
```

```python
pendiente = np.random.uniform(0.1,0.9,100)
for i in range (100):
  print(np.linalg.norm(moscu-(pendiente[i]*x+b)))
```

```python
norma= np.array([])
pendiente = np.random.uniform(0.1,0.9,100)
for i in range (100):
  norma = np.append(norma,np.linalg.norm(moscu-(pendiente[i]*x+b)))
```
